# Clustering Alanine Dipeptide using ShapeGMM and Meta Dynamics Weights

Alanine dipeptide (ADP) is a molecule that consists of 22 atoms.  It is a standard in the field of MD simulations in part because the major conformational changes of the molecule are well represented by the rotations around the $\phi$ and $\psi$ dihedral angles.  In the gas phase (and the force-field chosen here), the molecule populates three ($\phi$, $\psi$) states: C5 $\sim(-150,150)$, C7eq $\sim(-90,50)$, and minor state C7ax $\sim(50,-50)$.

Here we use a gas phase simulation run using well-tempered metadynamics (MetaD) with the collective variables being the $\phi$ and $\psi$ dihedral angles.  

When performing shapeGMM clustering, you must select the atoms you want to cluster.  In this case, since we are confident that the major conformational changes are in $\phi$ and $\psi$ dihedral angles, we select only the atoms that contribute to these angles.  Namely atoms: C, N, and CA.  Note that you can perform shapeGMM with all of the atoms of this molecule and you find similar behavior/clusterings.

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import MDAnalysis as md
from shapeGMMTorch import torch_sgmm
import torch
import nglview as nv
import pickle
import adp_scripts
```

## Read trajectory

We start by loading the trajectory.  I use `MDAnalysis` to read the trajectory file but then store the coordinates of my selected atoms in a numpy array of size `n_frames x n_atoms x 3`.  This is the input required for shapeGMM.

```python
prmtopFileName = "run_ala2_metad_sigma0.3_height1.2_pace500_bf10.gro"
trajFileName = "run_ala2_metad_sigma0.3_height1.2_pace500_bf10_wrapped.trr"
coord = md.Universe(prmtopFileName,trajFileName)
print("Number of atoms in trajectory:", coord.atoms.n_atoms)
print("Number of frames in trajectory:", coord.trajectory.n_frames)
# make atom selection
atomSel = coord.select_atoms('name C N CA')
print("Number of atoms in selection:", atomSel.n_atoms)
# create traj data of selection
traj_data = np.empty((coord.trajectory.n_frames,atomSel.n_atoms,3),dtype=float)
#loop traj
for ts in coord.trajectory:
    traj_data[ts.frame,:] = atomSel.positions
```

We can visualize the trajectory of the selected atoms using the `nglview` package

```python
view = nv.show_mdanalysis(coord)
view
```

## Load MetaD Weights

This trajectory is run with an active MetaD bias and thus the frames are not uniformly weighted.  Here I load the rbias value from the COLVARs file and convert it into a frame weight.

```python
rbias_weights = np.exp(np.loadtxt("run_ala2_metad_sigma0.3_height1.2_pace500_bf10.colvars.dat",comments="#")[:,4]/0.593)
rbias_weights /= np.sum(rbias_weights)
```

## Single shapeGMM: Comparing "kronecker" and "uniform" covariance models

We now have all of the data necessary to perform shapeGMM clustering.  I need to choose the number of clusters but in this case we have previous knowledge that it should be 3 so I will go with that.  Within shapeGMM, however, there is an additional option to choose either a `uniform` or `kronecker` covariance model.  I will do both and compare and contrast the outputs.

```python
dtype = torch.float64
device = torch.device("cpu")
sgmm_3clusters_uniform = torch_sgmm.ShapeGMMTorch(3,covar_type='uniform', device=device, dtype=dtype,verbose=True)
sgmm_3clusters_uniform.fit(traj_data,frame_weights=rbias_weights)
```

```python
dtype = torch.float64
device = torch.device("cpu")
sgmm_3clusters_kronecker = torch_sgmm.ShapeGMMTorch(3,covar_type='kronecker', max_steps=30, device=device, dtype=dtype, verbose=True)
sgmm_3clusters_kronecker.fit(traj_data, frame_weights=rbias_weights)
```

```python
# compute phi psi
from MDAnalysis.analysis import dihedrals
protein = coord.select_atoms("protein")
rama = dihedrals.Ramachandran(protein).run()
```

```python
# plot phi/psi scatter plot with cluster colorings
phi = rama.results.angles[:,0,0]
psi = rama.results.angles[:,0,1]
fig, ax = plt.subplots(1,2,figsize=(8,4),dpi=80)
c = sgmm_3clusters_uniform.cluster_ids
adp_scripts.rama_plot(ax[0],phi,psi,c,"Uniform covariance")
c = sgmm_3clusters_kronecker.cluster_ids
adp_scripts.rama_plot(ax[1],phi,psi,c,"Kronecker covariance")

#show plot
plt.tight_layout()
plt.show()
```

## Running shapeGMM a bunch of times...

ShapeGMM clustering, like any other EM GMM alogorithm, is heavily dependent on initialization.  It is likely to get caught in local maxima.  To alleviate this, we run shapeGMM a bunch of times on the same training set and pick the object with the maximum likelihood from these attempts.  ShapeGMM has a built in script to do this: `sgmm_fit_with_attempts`.  

This part of the notebook can take a while to run but I have saved final objects so you can just look at the output of the next few code snippets and then load the objects in a later code snippet.

```python
from shapeGMMTorch import scripts
```

```python
n_clusters = 3
n_attempts = 10
covar_type = "uniform"
sgmm_3clusters_uniform = scripts.sgmm_fit_with_attempts(traj_data, n_clusters, n_attempts, rbias_weights, covar_type=covar_type, device=device,dtype=dtype)
```

```python
# save object as a pickle file
fileObj = open('sgmm_3clusters_uniform.pickle', 'wb')
pickle.dump(sgmm_3clusters_uniform,fileObj)
fileObj.close()
```

```python
n_clusters = 3
n_attempts = 10
covar_type = "kronecker"
sgmm_3clusters_kronecker = scripts.sgmm_fit_with_attempts(traj_data, n_clusters, n_attempts, rbias_weights, covar_type=covar_type, device=device,dtype=dtype)
```

## As generative models

```python
def generate_md_analysis_object(sgmm,atomSel,n_frames=100000):
    # generate shapeGMM trajectory
    gen_traj = sgmm.generate(n_frames)
    # create MDAnalysis universe
    u = md.Universe.empty(sgmm.n_atoms, 1, atom_resindex=np.zeros(sgmm.n_atoms), trajectory=True)
    u.trajectory.n_frames = n_frames
    u.add_TopologyAttr('name', atomSel.names)
    u.add_TopologyAttr('type', atomSel.types)
    #u.atoms.names = atomSel.names
    #u.atom.types = atomSel.types
    sel_all = u.select_atoms("all")    
    # write pdb of frame 1 to use a ref
    pdb_file_name = "ref.pdb"
    atomSel.write(pdb_file_name)
    # write trajectory
    dcd_file_name = "sgmm_" + str(n_frames) + "frames.dcd"
    # write dcd of generated trajectory
    with md.Writer(dcd_file_name, sel_all.n_atoms) as W:
        for ts in range(n_frames):
            sel_all.positions = gen_traj[ts]
            W.write(sel_all)
    W.close()
    # load  MDAnalysis universe
    coord = md.Universe(pdb_file_name,dcd_file_name)
    return coord
```

```python
# generate uniform covariance trajectory
uniform_gen_traj = sgmm_3clusters_uniform.generate(1000000)
from MDAnalysis.lib.distances import calc_dihedrals
uniform_phis = calc_dihedrals(uniform_gen_traj[:,0,:],uniform_gen_traj[:,1,:],uniform_gen_traj[:,2,:],uniform_gen_traj[:,3,:])
uniform_psis = calc_dihedrals(uniform_gen_traj[:,1,:],uniform_gen_traj[:,2,:],uniform_gen_traj[:,3,:],uniform_gen_traj[:,4,:])
```

```python
# generate uniform covariance trajectory
kronecker_gen_traj = sgmm_3clusters_kronecker.generate(1000000)
kronecker_phis = calc_dihedrals(kronecker_gen_traj[:,0,:],kronecker_gen_traj[:,1,:],kronecker_gen_traj[:,2,:],kronecker_gen_traj[:,3,:])
kronecker_psis = calc_dihedrals(kronecker_gen_traj[:,1,:],kronecker_gen_traj[:,2,:],kronecker_gen_traj[:,3,:],kronecker_gen_traj[:,4,:])
```

```python
fig, ax = plt.subplots(1,3,figsize=(12,4),dpi=80)
contour_levels = np.arange(0,25,1)
# metaD FE
rbias = np.loadtxt("run_ala2_metad_sigma0.3_height1.2_pace500_bf10.colvars.dat",comments="#")[:,4]
adp_scripts.plot_rama_metaD_fe(ax[0],fig,phi,psi,rbias,"metaD FE",contour_levels=contour_levels)
# 2 clusters
adp_scripts.plot_rama_metaD_fe(ax[1],fig,uniform_phis*180.0/np.pi,uniform_psis*180.0/np.pi,np.ones(1000000),"SGMM Uniform Covar",contour_levels=contour_levels)
# 3 clusters
adp_scripts.plot_rama_metaD_fe(ax[2],fig,kronecker_phis*180.0/np.pi,kronecker_psis*180.0/np.pi,np.ones(1000000),"SGMM Kronecker Covar",contour_levels=contour_levels)
#show plot
plt.tight_layout()
plt.show()
```

# Cluster Scan

If we don't know the number of clusters *a priori* we perform a cluster scan and look at the log likelihood as a function of number of clusters.  We look for the ***elbow***, or minimum in the second derivative, of this plot.  ShapeGMM has built in scripts to perform a cross-validation cluster scan and to plot the resulting data.  

The cross validation script is contained in the `scripts` library and is titled: `cross_validate_cluster_scan`

The plotting script is contained in the `plots` library and is titled: `plot_log_likelihood_with_dd`

**NOTE:** This will take a while so unless you really want to perform the scan skip this and load the data.

```python
# define cluster array of number of clusters you want to test
cluster_array = np.arange(1,6,1).astype(int)
```

```python
# CV scan with non-uniform weights
training_set_size = 10000
n_training_sets = 3
n_attempts_per_training_set_and_n_clusters = 5
metaD_train, metaD_cv = scripts.cross_validate_cluster_scan(traj_data, training_set_size, frame_weights = rbias_weights, covar_type="kronecker", cluster_array = cluster_array, n_training_sets=n_training_sets, n_attempts = n_attempts_per_training_set_and_n_clusters, dtype=dtype, device=device)
```

### load log likelihood data for quick use

```python
# save data 
np.savetxt("kronecker_train.dat", metaD_train)
np.savetxt("kronecker_cv.dat", metaD_cv)
```
